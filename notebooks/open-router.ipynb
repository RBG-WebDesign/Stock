```python
import os
import sys
from pathlib import Path

# 1. Add project root to sys.path to allow imports from src/
project_root = Path().resolve().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

import asyncio
from dotenv import load_dotenv
from tqa.llm.openrouter import OpenRouterClient

# 2. Load environment variables (OPENROUTER_API_KEY)
load_dotenv(project_root / ".env")

async def test_openrouter_connectivity():
    print(f"Initializing OpenRouterClient...")
    
    async with OpenRouterClient() as client:
        print(f"Testing connectivity with model: {client.model}")
        
        # Simple test payload bypasses the ticker-specific logic for a basic health check
        payload = {
            "model": "inclusionai/ring-2.6-1t:free",
            "messages": [
                {"role": "user", "content": "Hello! This is a connectivity test from the Techno-Quantamental Analyzer. Please reply with 'CONNECTION SUCCESSFUL' if you can read this."}
            ]
        }
        
        print("Sending request to OpenRouter...")
        async with client._session.post(client.base_url, headers=client.headers, json=payload) as response:
            if response.status == 200:
                result = await response.json()
                content = result.get("choices", [{}])[0].get("message", {}).get("content", "")
                print(f"\n--- Response from LLM ---\n{content}\n-------------------------")
                print("\n✅ Connectivity Test Passed!")
            else:
                error_text = await response.text()
                print(f"❌ API Error {response.status}: {error_text}")

# 3. Run the test (Jupyter supports top-level await)
await test_openrouter_connectivity()
```
